Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)

1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)

```python
plt.figure(figsize=(8, 6))
colors = ['blue', 'gray', 'olive', 'brown', 'lightblue']
top_ciudades = df.nlargest(5, 'population_total')
plt.bar(top_ciudades['city'], top_ciudades['population_total'], color=colors, alpha=0.5)
for i, (ciudad, poblacion) in enumerate(zip(top_ciudades['city'], top_ciudades['population_total'])):
    plt.text(ciudad, poblacion, f"{poblacion:,}", ha='center', va='bottom')
plt.xlabel("Ciudades")
plt.ylabel('Habitantes (miles)')
plt.title('Ciudades más pobladas de California')
plt.xticks(rotation=360) #Poner las letras en horizontal
plt.yticks(range(0, 6000, 1000))  # Establecer los índices del eje y
plt.tight_layout()
plt.show()
```

2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)


```python
plt.figure(figsize=(8, 6))
plt.hist(df['area_total_km2'], bins=50, color='steelblue', label = "Hist Area")
plt.xlabel('Superficie (km²)', fontname = "Bradley Hand itc", fontsize=18)
plt.ylabel('Número de ciudades', fontname = "Bradley Hand itc", fontsize =18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.legend()
plt.show()
```

3. El scatter de área de agua de un ciudad con su poblacion.


```python
plt.figure(figsize=(8, 6))
plt.scatter(df['area_water_km2'], df['population_total'], color='green', alpha=0.5)
plt.xlabel('Área de agua en la ciudad (km²)')
plt.ylabel('Población (miles habitantes)')
plt.title('Población por kilómetro cuadrado de agua')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.show()
```

4. El boxplot de la altura de las ciudades.

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['elevation_m'], vert = False)
plt.xlabel('Altitud (m)')
plt.title('Boxplot de la Altura de las Ciudades')
plt.xticks(range(0, 2501, 500), fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout() #Ayuda a evitar que los elementos del gráfico se superpongan o se corten
plt.show()
```

```python
# Crear una figura con 2 filas y 2 columnas para los subgráficos
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Primer gráfico en la posición 0,0
colors = ['blue', 'gray', 'olive', 'brown', 'lightblue']
top_ciudades = df.nlargest(5, 'population_total')
axs[0, 0].bar(top_ciudades['city'], top_ciudades['population_total'], color=colors, alpha=0.5)
for i, (ciudad, poblacion) in enumerate(zip(top_ciudades['city'], top_ciudades['population_total'])):
    axs[0, 0].text(ciudad, poblacion, f"{poblacion:,}", ha='center', va='bottom')
axs[0, 0].set_xlabel("Ciudades")
axs[0, 0].set_ylabel('Habitantes (miles)')
axs[0, 0].set_title('Ciudades más pobladas de California')
axs[0, 0].tick_params(axis='x', rotation=360) # Rotar las etiquetas del eje x

# Segundo gráfico: en la posición 0,1
axs[0, 1].hist(df['area_total_km2'], bins=50, color='steelblue', label="Hist Area")
axs[0, 1].set_xlabel('Superficie (km²)',  fontname = "Bradley Hand itc", fontsize=18)
axs[0, 1].set_ylabel('Número de ciudades',  fontname = "Bradley Hand itc", fontsize=18)
axs[0, 1].set_title('Histograma del Área Total de las Ciudades')

# Tercer gráfico: en la posición 1, 0
axs[1, 0].scatter(df['area_water_km2'], df['population_total'], color='green', alpha=0.5)
axs[1, 0].set_xlabel('Área de agua en la ciudad (km²)')
axs[1, 0].set_ylabel('Población (miles habitantes)')
axs[1, 0].set_title('Población por kilómetro cuadrado de agua')

# Cuarto gráfico: en la posición 1,1
axs[1, 1].boxplot(df['elevation_m'], vert=False, patch_artist=True, boxprops=dict(facecolor="red", color="red"), whiskerprops=dict(color="red"))
axs[1, 1].set_xlabel('Altitud (m)')
axs[1, 1].set_title('Boxplot de la Altura de las Ciudades')

# Ajustar el diseño y mostrar la figura
plt.tight_layout()
plt.show()
```
