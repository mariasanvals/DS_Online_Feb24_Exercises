1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)

```python
plt.figure(figsize=(8, 6))
colors = ['blue', 'gray', 'olive', 'brown', 'lightblue']
top_ciudades = df.nlargest(5, 'population_total')
plt.bar(top_ciudades['city'], top_ciudades['population_total'], color=colors, alpha=0.5)
for i, (ciudad, poblacion) in enumerate(zip(top_ciudades['city'], top_ciudades['population_total'])):
    plt.text(ciudad, poblacion, f"{poblacion:,}", ha='center', va='bottom')
plt.xlabel("Ciudades")
plt.ylabel('Habitantes (miles)')
plt.title('Ciudades ms pobladas de California')
plt.xticks(rotation=360) #Poner las letras en horizontal
plt.yticks(range(0, 6000, 1000))  # Establecer los ndices del eje y
plt.tight_layout()
plt.show()
```

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)


```python
plt.figure(figsize=(8, 6))
plt.hist(df['area_total_km2'], bins=50, color='steelblue', label = "Hist Area")
plt.xlabel('Superficie (km)', fontname = "Bradley Hand itc", fontsize=18)
plt.ylabel('Nmero de ciudades', fontname = "Bradley Hand itc", fontsize =18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.legend()
plt.show()
```

3. El scatter de rea de agua de un ciudad con su poblacion.


```python
plt.figure(figsize=(8, 6))
plt.scatter(df['area_water_km2'], df['population_total'], color='green', alpha=0.5)
plt.xlabel('rea de agua en la ciudad (km)')
plt.ylabel('Poblacin (miles habitantes)')
plt.title('Poblacin por kilmetro cuadrado de agua')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.show()
```

4. El boxplot de la altura de las ciudades.

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['elevation_m'], vert = False)
plt.xlabel('Altitud (m)')
plt.title('Boxplot de la Altura de las Ciudades')
plt.xticks(range(0, 2501, 500), fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout() #Ayuda a evitar que los elementos del grfico se superpongan o se corten
plt.show()
```

```python
# Crear una figura con 2 filas y 2 columnas para los subgrficos
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Primer grfico en la posicin 0,0
colors = ['blue', 'gray', 'olive', 'brown', 'lightblue']
top_ciudades = df.nlargest(5, 'population_total')
axs[0, 0].bar(top_ciudades['city'], top_ciudades['population_total'], color=colors, alpha=0.5)
for i, (ciudad, poblacion) in enumerate(zip(top_ciudades['city'], top_ciudades['population_total'])):
    axs[0, 0].text(ciudad, poblacion, f"{poblacion:,}", ha='center', va='bottom')
axs[0, 0].set_xlabel("Ciudades")
axs[0, 0].set_ylabel('Habitantes (miles)')
axs[0, 0].set_title('Ciudades ms pobladas de California')
axs[0, 0].tick_params(axis='x', rotation=360) # Rotar las etiquetas del eje x

# Segundo grfico: en la posicin 0,1
axs[0, 1].hist(df['area_total_km2'], bins=50, color='steelblue', label="Hist Area")
axs[0, 1].set_xlabel('Superficie (km)',  fontname = "Bradley Hand itc", fontsize=18)
axs[0, 1].set_ylabel('Nmero de ciudades',  fontname = "Bradley Hand itc", fontsize=18)
axs[0, 1].set_title('Histograma del rea Total de las Ciudades')

# Tercer grfico: en la posicin 1, 0
axs[1, 0].scatter(df['area_water_km2'], df['population_total'], color='green', alpha=0.5)
axs[1, 0].set_xlabel('rea de agua en la ciudad (km)')
axs[1, 0].set_ylabel('Poblacin (miles habitantes)')
axs[1, 0].set_title('Poblacin por kilmetro cuadrado de agua')

# Cuarto grfico: en la posicin 1,1
axs[1, 1].boxplot(df['elevation_m'], vert=False, patch_artist=True, boxprops=dict(facecolor="red", color="red"), whiskerprops=dict(color="red"))
axs[1, 1].set_xlabel('Altitud (m)')
axs[1, 1].set_title('Boxplot de la Altura de las Ciudades')

# Ajustar el diseo y mostrar la figura
plt.tight_layout()
plt.show()
```
