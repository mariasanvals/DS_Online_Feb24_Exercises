### #1.2

Crea una función similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una función para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numérica del dataframe
3. Una columna numérica del dataframe diferente a la anterior.
4. Un columna categórica (Opcional, valor por defecto None)
5. Un valor para el tamaño de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamaño
6. Un sexto argumento "escala", que permita manipular la escala del tamaño en el caso de que el quinto argumento sea una cadena. Este argumento multiplicará a los valores de la columna indicada en el parámetro anterior (así si quiero dividir por 10000, este sexto argumento tendrá que valer 1/10000)

La función debe pintar el diagrama de dispersión de la columna numérica 1, en el eje X, frente a la columna numérica 2, en el eje Y, usando un código de colores en función de la columna categórica y con el tamaño de los puntos según el argumento correspondiente. 

Si no se diera una columna categórica como argumento, el color no debe pasarse como parámetro a la función de Seaborn o Matplotlib que uses.

EXTRA: Añade un argumento que controle si la función pinta una leyenda para colores y tamaños o un colorbar y una leyenda para los tamaños. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesión en vivo)

```python
def diagrama_dispersion (df, col_num1, col_num2, col_cat = None, size= "", scale = 1):
    plt.figure(figsize=(17,10))
    if isinstance(size, str):  # Si el tamaño es una columna del DataFrame
        size_values = size * scale  # Aplicar la escala si es necesario
        sns.scatterplot (data = df, x = col_num1, y= col_num2, hue= col_cat, s=size_values, palette= "viridis", alpha = 0.5)
    else:
        sns.scatterplot (data = df, x = col_num1, y= col_num2, s=size, alpha = 0.5)
    plt.legend (loc="upper right")
    plt.show()
```

```python
#Prueba de función
diagrama_dispersion(data, 'area_total_km2', 'population_total')

# Llamada a la función con color por categoría y tamaño de puntos personalizado
diagrama_dispersion(data, 'area_total_km2', 'population_total', cat_col='city', size='elevation_m', scale=0.1)
```

```python
df_titanic = pd.read_csv("./data/titanic.csv")
df_titanic.head()
```

```python
# 1- El titanic fue un accidente con una tasa de mortalidad alta.  

# frecuencias absolutas
categorias = df_titanic.alive.value_counts().index.to_list()
frecuencias = df_titanic.alive.value_counts().to_list()
# frecuencias  relativas
frecuencias_relativas = (df_titanic.alive.value_counts(normalize = True) * 100).to_list()

plt.figure(figsize=(10,10))
plt.bar(categorias, frecuencias_relativas);
```

```python
# 2- En esa ocasión los datos apoyan que se dio el "Las mujeres y los niños primero", pero...  

sns.catplot( x = "alive",
            hue = "who", #diferencias color por clase
            kind = "count",#permite diferentes gráficos
            edgecolor = ".6",
            orient = "v",
            data = df_titanic.sort_values("who"));
```

```python
tabla_contingencia_who = pd.crosstab(df_titanic["alive"],df_titanic["who"])
tabla_contingencia_who

```

```python
cat_colx = "alive"
cat_coly = "who"
frecuencias_absolutas = df_titanic.groupby(cat_coly, as_index = False)[cat_colx].value_counts()
frecuencias_relativas = df_titanic.groupby(cat_coly, as_index = False)[cat_colx].value_counts(normalize = True)
print(frecuencias_absolutas)
print(frecuencias_relativas)
```

```python
tabla_contingencia = pd.crosstab(df_titanic["alive"],df_titanic["who"]).T
tabla_contingencia["porcentaje_supervivencia"] = (tabla_contingencia["yes"]/(tabla_contingencia["yes"] + tabla_contingencia["no"])).round(3) * 100
tabla_contingencia
```

```python
# 3- ... que la clase en la que se viajes tuvo una influencia significativa y podría explicar...  

sns.catplot( x = "alive",
            hue = "class", #diferencias color por clase
            kind = "count",#permite diferentes gráficos
            edgecolor = ".6",
            orient = "v",
            data = df_titanic.sort_values("class"));
```

```python
tabla_contingencia_class = pd.crosstab(df_titanic["alive"],df_titanic["class"])
tabla_contingencia_class
```

```python
cat_colx = "alive"
cat_coly = "class"
frecuencias_absolutas = df_titanic.groupby(cat_coly, as_index = False)[cat_colx].value_counts()
frecuencias_relativas = df_titanic.groupby(cat_coly, as_index = False)[cat_colx].value_counts(normalize = True)
print(frecuencias_absolutas)
print(frecuencias_relativas)
```

```python
# 4- ... las diferencias de tasas de supervivencia entre mujeres y niños.    

# Filtrar los datos para obtener solo niños y mujeres
niños = df_titanic[df_titanic['who'] == "child"]
mujeres = df_titanic[df_titanic['who'] == "woman"]

# Crear dos gráficos de barras separados, uno para niños y otro para mujeres
fig, axes = plt.subplots(2, 1, figsize=(10, 10))

sns.countplot(data=niños, x='class', hue='alive', palette='coolwarm', ax=axes[0])
axes[0].set_title('Sobrevivientes y fallecidos por clase - Niños')
axes[0].set_xlabel('Clase')
axes[0].set_ylabel('Número de pasajeros')
axes[0].legend(title='Sobrevivió', loc='upper right', labels=['No', 'Sí'])

sns.countplot(data=mujeres, x='class', hue='alive', palette='coolwarm', ax=axes[1])
axes[1].set_title('Sobrevivientes y fallecidos por clase - Mujeres')
axes[1].set_xlabel('Clase')
axes[1].set_ylabel('Número de pasajeros')
axes[1].legend(title='Sobrevivió', loc='upper right', labels=['No', 'Sí'])

plt.tight_layout()
plt.show()
```
